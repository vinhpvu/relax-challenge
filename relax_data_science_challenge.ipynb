I had to create 'adopted_user' in order to define adopted users. Adopted user is as a user who has logged into the product on three different days.


Interesting findings

•	Adopted User is imbalanced; it has 88% of the users are not adopted users and 12% of the users are adopted users

•	Source Creation - Majority of the users were invited to an organization

•	January to May, it has the most of non-adopted users registered

•	Account Creation by month May has the most adopted user, and it has the least non-adopted user  


Most Interesting Finding:
    
•	User email distribution- Gmail has the most adopted user, and yahoo has the most non-adopted user

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import datetime as dt
%matplotlib inline
```

```python
users = pd.read_csv('takehome_users.csv', encoding='latin-1')
users.head()
```

```python
eng = pd.read_csv('takehome_user_engagement.csv')
eng.head(5)
```

```python
visits = eng["user_id"].value_counts()
adopted = visits[visits >= 3]
```

```python
adopted
```

```python
for index, val in adopted.iteritems(): 
    users.loc[users['object_id'] == index, 'adopted_user'] = 1
```

```python
users.loc[users['object_id'] == 2]
```

```python
users.head()
```

```python
users=users.fillna(0)
users.head()
```

# Adopted users

Summary:

About 81% of the users are not adopted users

About 19% of the users are adopted users

```python
users.adopted_user.value_counts(True)
```

# Creation Source Distribution

Majority of the users were invited to an organization.


```python
f, ax = plt.subplots(figsize=(15, 4))
sns.countplot(y="creation_source", hue='adopted_user', data=users).set_title('Creation Source Distribution');
```

# Creation Time Distribution

```python
users['creation_time'] = pd.to_datetime(users.creation_time)
```

```python
users['year'] = users['creation_time'].dt.year
users['month'] = users['creation_time'].dt.month
users['day'] = users['creation_time'].dt.day
```

```python
users.email = users.email.apply(lambda x: x.split('@')[1].split('.')[0])
```

```python
main_domain = ['gmail', 'yahoo', 'jourrapide', 'cuvox', 'gustr', 'hotmail']
users.email[~users.email.isin(main_domain)] = 'others'
```
